```python
min = sorted_df2.min()
max = sorted_df2.max()
mean = sorted_df2.mean()
median = sorted_df2.median()
disp = sorted_df2.var()
std = sorted_df2.std()
q1 = sorted_df2.quantile(0.25)
q3 = sorted_df2.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
gap = sorted_df2.isna().sum()
sorted_df3 = pd.concat([min, max, mean, median, disp, std, iqr, gap], axis=1)
sorted_df3.columns = ['min', 'max', 'mean', 'median', 'disp', 'std', 'iqr', 'gap']
print(sorted_df3)
```

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

surgery? -> удалить строку с неопределенным значением (жертвуем единичным значением)
Age - норм
rectal temperature -> проставить медиану (средняя температура для большой выборки)
pulse -> проставить медиану (средний пульс для большой выборки)
respiratory rate -> заполнить нолями (неинформативное поле по описанию)
temperature of extremities -> заполнить нолями (категорийные данные, невозможно предсказать значение)
pain -> заполнить нолями (категорийные данные, невозможно предсказать значение)
outcome -> удалить строку с неопределенным значением (жертвуем единичным значением)

```python
sorted_df2.dropna(subset=['surgery?'])
sorted_df2.dropna(subset=['outcome'])
sorted_df2 = sorted_df2.reset_index(drop=True)
sorted_df2['rectal temperature'] = sorted_df2['rectal temperature'].fillna(sorted_df2['rectal temperature'].median())
sorted_df2['pulse'] = sorted_df2['pulse'].fillna(sorted_df2['pulse'].median())
sorted_df2['respiratory rate'] = sorted_df2['respiratory rate'].fillna(0)
sorted_df2['temperature of extremities'] = sorted_df2['temperature of extremities'].fillna(0)
sorted_df2['pain'] = sorted_df2['pain'].fillna(0)
```
