respiratory rate -> заполнить нолями (неинформативное поле по описанию, широкий разброс значений)
outcome -> удалить строку с неопределенным значением (жертвуем единичным значением)

```python
sorted_df.dropna(subset=['surgery?'], inplace=True)
sorted_df.dropna(subset=['outcome'], inplace=True)
sorted_df['respiratory rate'] = sorted_df['respiratory rate'].fillna(0)
sorted_df = sorted_df.reset_index(drop=True)
```

```python
sorted_df.pivot_table(index = 'surgery?', aggfunc = 'median')
```

у оперированных лошадей (surgery?=1) pain и temperature of extremities выше;

```python
sorted_df.pivot_table(index = ['Age', 'surgery?'], aggfunc = 'median')
```

есть разброс медианы (pulse, ) в зависимости от возраста и проведения операции, но не для rectal temperature

rectal temperature -> проставить медиану (средняя температура для большой выборки)
pulse -> проставить медиану (средний пульс для большой выборки)
temperature of extremities -> заполнить модой (категорийные данные)
pain -> заполнить модой (категорийные данные)

```python
sorted_df['rectal temperature'] = sorted_df['rectal temperature'].fillna(sorted_df['rectal temperature'].median())
sorted_df['pulse'] = sorted_df['pulse'].fillna(sorted_df.groupby(['Age', 'surgery?'])['pulse'].transform('median'))
sorted_df['temperature of extremities'] = sorted_df['temperature of extremities'].fillna(sorted_df.groupby(['Age', 'surgery?'])['temperature of extremities'].transform(lambda x: x.mode()[0]))
sorted_df['pain'] = sorted_df['pain'].fillna(sorted_df.groupby(['Age', 'surgery?'])['pain'].transform(lambda x: x.mode()[0]))
sorted_df.head(20)
```
