Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

оценка 2 и ниже — низкий рейтинг;
оценка 4 и ниже — средний рейтинг;
оценка 4.5 и 5 — высокий рейтинг.
Результат классификации запишите в столбец class.

```python
import pandas as pd
def class_def(num):
    if (num > 0 and num <= 2): result = 'низкий рейтинг'
    elif (num > 2 and num <= 4): result = 'средний рейтинг'
    elif (num == 4.5 or num == 5): result = 'высокий рейтинг'
    else: result = 'неопределенный рейтинг'
    return result
df = pd.read_csv('ratings.csv', sep = ",")
df = df.groupby('movieId').mean().reset_index()
df['class'] = df['rating'].apply(class_def)
df.head()
```

Задание 2

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'] }

Результат классификации запишите в отдельный столбец region.

```python
import pandas as pd
data = pd.read_csv('keywords.csv')
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
def region_def(keyword):
    list_words = keyword.split()
    for i in list_words:
        for key in geo_data:
            for items in geo_data[key]:
                if items == i: return key
    return 'undefined'
data['region'] = data['keyword'].apply(region_def)
data.head(790)
```

```python
import pandas as pd
ratings = pd.read_csv('ml-1m/ratings.dat', sep = "::", encoding='utf8', names = ["UserID", "MovieID", "Rating", "Timestamp"])
movies = pd.read_csv('ml-1m/movies.dat', encoding = 'latin-1', sep = "::", names = ["MovieID", "Title", "Genres"])
years = list(range(1950,2011))
def production_year(movie):
    for i in years:
        if str(i) in movie: return i
    return 1900
movies['Year'] = movies['Title'].apply(production_year)
ratings = ratings.groupby('MovieID').mean().reset_index()
overall = pd.merge(ratings, movies, on='MovieID')
overall.groupby('Year')[['Rating']].mean().reset_index().sort_values('Rating', ascending=False)[:10]
```
