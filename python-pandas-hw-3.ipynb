Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

если источник traffic_source равен Yandex или Google, то в source_type ставится organic;
для источников paid и email из России ставим ad;
для источников paid и email не из России ставим other;
все остальные варианты берём из traffic_source без изменений.


```python
import pandas as pd
data = pd.read_csv('visit_log.csv', sep=';')
data.head()
```

```python
import pandas as pd
def filter_data(reg, source):
    if (source == 'yandex' or source == 'google'): return 'organic'
    elif (source =='paid' and reg == 'Russia'): return 'ad'
    elif (source =='paid' and reg != 'Russia'): return 'other'
    else: return source
data = pd.read_csv('visit_log.csv', sep=';')
data['source_type'] = data.apply(lambda x: filter_data(x['region'], x['traffic_source']), axis =  1)
data.head()
```

Задание 3
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
import pandas as pd
data = pd.read_csv('ratings.dat', sep = "::", names = ["UserID", "MovieID", "Rating", "Timestamp"])
data.head()
```

```python
import pandas as pd
data = pd.read_csv('ratings.csv', sep = ",")
df1 = data.groupby('userId')['rating'].count()
agg_func_selection = {'timestamp': ['first', 'last']}
df2 = data.sort_values(by=['timestamp'], ascending=False).groupby(['userId']).agg(agg_func_selection)
df3 = pd.concat([df1, df2], axis=1).reindex(df1.index)
df4 = df3[df3['rating'] > 100]
df4.head()
```
