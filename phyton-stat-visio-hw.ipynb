```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('2019.csv', sep=',')
data.head()

```

```python
plt.plot(data['Country or region'],data['Score'])
plt.xticks(rotation=90, fontsize=5)
plt.figure(figsize=(500,500))
```

```python
data['Score'].plot(kind='hist', 
                   title='Оценка счастья', 
                    xlabel='Шкала счастья от 0 до 10',
                    ylabel='Частота ответов',
                   bins=40)
plt.grid()
```

В основном ответы находятся в среднем диапазоне: 4-6,5

```python
data['Social support'].plot(kind='hist', 
                     alpha = 0.5,
                     bins=20,
                     density = True)


data['GDP per capita'].plot(kind='hist', 
                     alpha = 0.5,
                     bins=20,
                     density = True)
plt.legend(loc='upper left')
plt.grid()
plt.title('Сравнение распределений социальной поддержки и ВВП на человека')
```

Расходы на социальную поддержку тем больше, чем больше уровень ВВП на человека

```python
pie_data = round(data['Social support'], 1)
pie_data = pie_data.value_counts()
#pie_data = pie_data.sort_values()
pie_data.plot(kind='pie', 
                title='Распределение стран по величине уровня социальной поддержки',
                autopct='%1.1f%%'
                )
```

Большинство стран (71,8%) имеют высокий уровень социальной поддержки. Наивысший уровень соцподдержки имеют 3,8% стран.

```python
sns.pairplot(data[['Score', 'Social support', 'Healthy life expectancy','Freedom to make life choices']])
```

Уровень счастья зависит от уровня социальной поддержки в стране и продолжительности жизни, но не зависит от степени свободы в стране. 
Продолжительность жизни зависит от уровня счастья и уровня социальной поддержки и совсем не зависит от свободы.
