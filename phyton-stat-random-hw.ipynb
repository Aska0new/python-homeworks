Домашнее задание к лекции «Случайные события. Случайные величины»

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

*– любая цифра

Вопросы:
Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?

```python
from random import randint
```

```python
def winning(num):
    if num == 777: return 200
    elif num == 999: return 100
    elif num == 555: return 50
    elif num == 333: return 15
    elif num == 111: return 10
    elif num % 100 == 77: return 5
    elif num % 10 == 7: return 3
    elif num % 100 == 0: return 2
    elif num % 10 == 0: return 1
    else: return 0

def fell_out():
    return randint(0, 999)

spent_money = 0
received_money = 0
step = 0
while True:
    play = input('Введите 1 для игры. Введите 0 для выхода.')
    if play == '1': 
        step +=1
        spent_money += 1
        print(current := fell_out())
        gap = winning(current)
        if gap > 0: received_money = received_money + gap
        print('Попытка: ', step)
        print('Денег выигранно: ', received_money)
        print('Денег проигранно: ', spent_money)
    elif play == '0': break
    else: 
        print('Некорректное значение')
```

**Дополнительное задание повышенной сложности. **
Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

```python
def winning(num):
    if num == 777: return 200
    elif num == 999: return 100
    elif num == 555: return 50
    elif num == 333: return 15
    elif num == 111: return 10
    elif num % 100 == 77: return 5
    elif num % 10 == 7: return 3
    elif num % 100 == 0: return 2
    elif num % 10 == 0: return 1
    else: return 0

def fell_out():
    return randint(0, 999)

spent_money = 0
received_money = 0
win = 0
loss = 0
trials = input('Количество игр?')
for i in range(int(trials)):
    spent_money += 1
    random_score = fell_out()
    gap = winning(random_score)
    if gap > 0: 
        received_money = received_money + gap
        win += 1
    else: 
        spent_money += 1
        loss += 1
print('Денег выигранно: ', received_money)
print('Денег проигранно: ', spent_money)
print('Средний выигрыш//проигрыш: ', round((received_money-spent_money)/int(trials), 2))
print('Вероятность выигрыша: ', round(win/int(trials), 2))
print('Вероятность проигрыша: ', round(loss/int(trials), 2))
print('Теоретическая вероятность выигрыша: ', round(1/(10*10*10), 8))
```
